```python
# Classification of Movies Based on Ratings [Excellent, Good, and Average] 

bins = [0,5.0,7.0,10.0]
group_names = ['Average', 'Good', 'Excellent',]
imdb_movie_data['Movie Rating'] = pd.cut(imdb_movie_data['Rating'], bins, labels = group_names)
imdb_movie_data['Movie Rating']
Movie_Rating = imdb_movie_data['Movie Rating']
Movie_Title = imdb_movie_data['Title']
Movie_classification = pd.concat([Movie_Rating, Movie_Title], axis = 1)
Movie_classification
```
