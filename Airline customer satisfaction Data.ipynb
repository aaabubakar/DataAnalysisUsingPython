# Airline Passenger Satisfaction Data Cleaning and Analysis 

```python
#first i start by importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
#importing the csv file and mapping it to a dataframe
airline = pd.read_csv("Downloads\\airline_passenger_satisfaction.csv")
```

```python
#checking and viewing the data columns of the dataset 
airline.columns
```

```python
#using the shape function to check the number of rows and columns
airline.shape
```

```python
#looking at the brief data info
airline.info()
```

```python
#checking the data using the first five rows of the data
airline.head()
```

```python
#checking the last five rows of the data
airline.tail()
```

```python
#making/setting the ID as the index
airline = airline.set_index('ID')
airline
```

```python
#checking for duplicate values
airline[airline.duplicated()]
```

```python
#checking for null values
airline.isnull().sum()
```

```python
#Alternatively checking for null values using the na function
airline.isna()
```

###### it is hard to see the null values like this but we can add a sum function to isna function

```python
airline.isna().sum()
```

```python
#Alternatively
airline.isnull().sum()
```

```python
# alternatively we can check for empty data in the dataset using the heatmap plot procedure where a the black is the dataset and the white spaces are the empty of Nan Values 
sns.heatmap(airline.isnull(), cbar = False, cmap = 'magma')
plt.show()
```

```python
#removing the na Values 
airline = airline.dropna(axis=0)
```

```python
#verifying the removal of na values

sns.heatmap(airline.isnull(), cbar = False)
plt.show()
```

```python
airline.isnull().sum()
```

```python
#checking the descriptive statistics using the describe function
airline.describe()
```

```python
#checking for the statistical correlation of the data using the corr() function 
airline.corr()
```

```python
#using the heatmap to check the correlation graphically
sns.heatmap(airline.corr(), cbar = True);
```

```python
#Class Distribution
sns.histplot(airline['Class'], bins = 6)
plt.ylim(0,80000);
```

```python
#Type of Travel
sns.histplot(airline['Type of Travel'], bins = 3)

plt.ylim(0,100000);
```

```python
#Satisfaction count based on customer Type
airline.groupby(['Satisfaction', 'Customer Type'])['Gender'].count()
```

```python
#Type of travel based on Gender distribution
airline.groupby(["Type of Travel","Gender"])["Gender"].count()
```

```python
#Gender Value Count
airline["Gender"].value_counts()
```

```python
airline["Gender"].value_counts().plot(kind='bar')
plt.ylim(0, 80000)
plt.xlabel('Gender')
plt.ylabel("Count")
plt.title("Number of cumtomers by Gender");
```
